Below code will import data from csv file and put it on to a DataFrame **df

```python
import pandas as p
file = r'C:\Users\nvishwakarma\Documents\Python Scripts\HRAnalysis/HRData.csv'
df = p.read_csv(file)
df[-10:]
```

    ## Based on the department what are the mean of satisfaction_level
    

```python
df.groupby('sales').mean().sort_values('left', ascending=False)
```

 Separating data where employees have left vs still working

```python
df_StillWorking = df[df['left']!=1]
df_StillWorking[-5:]
```

```python
 df_StillWorking.shape
```

```python
df_left = df[df['left']==1]
df_left[-5:]
```

```python
df_left.shape
```

Highest and Lowest satisfaction among still working employees by Department

```python
df_grp_sales = df_StillWorking.groupby('sales')
df_grp_sales[['satisfaction_level']].mean().sort_values('satisfaction_level', ascending=False)
```

```python
df_left.groupby('sales')[['satisfaction_level']].mean().sort_values('satisfaction_level', ascending=True)#.size()
```

```python
df_StillWorking_grp_sales_salary = df_StillWorking.groupby(['sales', 'salary'])
df_StillWorking_grp_sales_salary.mean().sort_values('satisfaction_level', ascending=False)['satisfaction_level'].to_frame()
```
